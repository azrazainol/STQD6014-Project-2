# **Introduction**

Population age is an important factor that contributes to the wellbeing of a country. Additionally, fertility rate is considered important to ensure that the population age is within the working age. On average, a stable population is achieved by maintaining a fertility rate of 2.1 per woman (OECD, 2024). However, developed countries tend to show a pattern of declining fertility rates (Singh, 2023). The decline is due to various reasons such as smaller families having better family economics, gender inequality leading to women preferring single lives, and dissatisfactional marriages.

Low fertility rate is also linked with divorce. A country with liberal divorce laws causes a drop by 0.2 percent in fertility rate (Bellido, & Marcen, 2014). Finally, the mortality rate also influences the population number and the economic development of a country. Developed countries have seen a decrease in mortality rate (Freeman, et. al., 2020). A low mortality rate requires the labour of the working population so that the country may support the welfare of the older population. A low working population would result in insufficient funds to care for the elderly people.

In relation, South Korea, known for Samsung, Kimchi and its entertainment industry, Hallyu, has been the topic of rave on the internet the past few years. With the growing interest towards South Korea, its economy rapidly increased at an outstanding rate among other developed countries.

# **About the dataset:**

The dataset was employed from Kaggle: https://www.kaggle.com/datasets/alexandrepetit881234/korean-demographics-20002022

The dataset consists of data between January 2000 until June 2022 and has 12 variables:
- Date: The date by months
- Region: The state
- Birth: the number of births
- Birth_rate: The birth rate
- Death: The number of deaths
- Death_rate: The death rate
- Divorce: The number of divorce cases
- Divorce_rate: The divorce rate
- Marriage: The number of marriages
- Marriage_rate: The marriage rate
- Natural_growth: The difference between the number of births and the number of deaths
- Natural_growth_rate: The percentage of the natural growth against the population number

#### **Loading Necessary Packages**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

#### **Importing the Data and Data Exploration**

```python
# import data from url as pandas dataframe
url = "http://tinyurl.com/27273hz7"
demo = pd.read_csv(url)
demo.head(18)
```

```python
# Check the dataframe structure
print(demo.info())
```

```python
print(demo['Region'].unique())
```

```python
# Check if missing values are present in the dataframe
demo.isnull().sum()
```

#### **Data Cleaning**

```python
# Identify rows with missing values
rows_with_missing_values = demo[demo.isnull().any(axis=1)]
print(rows_with_missing_values)
print("\nRows with missing values:")
print(len(rows_with_missing_values))
```

```python
# Remove rows with missing values from the 'demo' because the values cannot be assumed using imputation methods
kdemo = demo.dropna()

# Display the DataFrame without missing values
print("DataFrame without Missing Values:")
print(kdemo)
```

```python
# Convert Date column to date format
kdemo['Date'] = pd.to_datetime(kdemo['Date'])
## Create a new column that shows the month and year only
## kdemo['Month'] = kdemo['Date'].dt.to_period('M')
# Create a new column that shows the year only
kdemo['Year'] = kdemo['Date'].dt.year
```

```python
print(kdemo.info())
```

```python
kdemo_states = kdemo[kdemo['Region'] != 'Whole country']
kdemo_states
```

```python
kdemo_country = kdemo[kdemo['Region'] == 'Whole country']
kdemo_country
```

# **Analysis**

#### **Data Visualisation Using Seaborn**

1. South Korea birth rate from 2000 until the first half of 2022
2. The overall percentage of births and birth rates by region between 2000 until the first half of 2022
3. The overall marriage and divorce rate of South Korea between 2000 until the first half of 2022
4. The relationship between the natural growth rate and marriage and divorce rate
5. The average death rate of South Korea between 2000 until the first half of 2022

```python
# Set the style of seaborn
sns.set(style="whitegrid")
```

```python
# 1. South Korea birth rate from 2000 until the first half of 2022
# by month
plt.figure(figsize=(15, 6))
sns.lineplot(x='Date', y='Birth_rate', data=kdemo_states)
plt.title('Birth Rate Over Time')
plt.show()
```

```python
# 1. South Korea Birthrate from 2000 until the first half of 2022
# by year
plt.figure(figsize=(15, 6))
sns.lineplot(x='Year', y='Birth_rate', data=kdemo_states)
plt.title('Birth Rate Over Time')
plt.show()
```

Based on both of the line plots above, a steep downwards trend can be seen between the year 2000 until 2005 and also the year 2015 until 2022. The overall pattern shows a decrease in the birth rate of South Korea.

The first steep decline is claimed to be due to the number of child regulations that the government implemented in order to control the birth rate. However, the second decline is claimed to be caused by the choice of South Korean women to be independent.

After examining the birth rate of South Korea, the regions contributing the most to the population number is examined using the number of births per region and birth rate of the regions

```python
# 2. The overall percentage of births and birth rates by region between 2000 until the first half of 2022
# Number of Births
birth_by_region = kdemo_states.groupby('Region')['Birth'].sum()
plt.figure(figsize=(15, 6))
plt.subplot(1, 2, 1)
plt.pie(birth_by_region, labels=birth_by_region.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('husl', n_colors=len(birth_by_region)))
plt.title('Birth Percentage by Region (2000-2022)')

# Birth Rate
birthrate_by_region = kdemo_states.groupby('Region')['Birth_rate'].sum()
plt.subplot(1, 2, 2)
plt.pie(birthrate_by_region, labels=birthrate_by_region.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('husl', n_colors=len(birthrate_by_region)))
plt.title('Birth Rate Percentage by Region (2000-2022)')

plt.tight_layout()
plt.show()
```

Based on the percentage of the number of births, it can be seen that Gyeonggi-do has the highest percentage of births followed by Seoul. Although Gyeonggi-do is still the highest in birth rate percentage, Seoul has a lower percentage of birthrate than half of the regions. Seoul is the capital of South Korea and is the most compact state of the country holding about 70% of the population Gyeonggi has despite being almost 10 times smaller than Gyeonggi. However, the HDI for Seoul and Gyeonggi-do are both 0.952 but both differ in ranking of birth rates among other states. Unlike what was found in past literature, there may be other underlying reasons to why the birth rate declines aside from being more developed.


```python
# 3. The overall marriage and divorce rate of South Korea between 2000 until the first half of 2022
# Marriage Rate
kdemo['Year'] = kdemo['Date'].dt.year
marriage_rate_by_year = kdemo.groupby('Year')['Marriage_rate'].mean()
plt.figure(figsize=(12, 6))
sns.barplot(x=marriage_rate_by_year.index, y=marriage_rate_by_year)
plt.title('Average Marriage Rate per Year')
plt.show()
```

It can be observed from the plot above that there are peaks in marriage rates in 2000 and 2007 with a slight dip in marriage rate where the lowest between those years was in 2003. In 2011 until 2014, the marriage rates did not have any drastic changes. Despite that, a constant decline was observed from 2015 until 2021 which is similar to the decline in birth rates earlier. This pattern may be indicating that there could be changes in standard of living or societal norms during the period.

```python
# 3. The overall marriage and divorce rate of South Korea between 2000 until the first half of 2022
# Divorce Rate
kdemo['Year'] = kdemo['Date'].dt.year
divorce_rate_by_year = kdemo.groupby('Year')['Divorce_rate'].mean()
plt.figure(figsize=(12, 6))
sns.barplot(x=divorce_rate_by_year.index, y=divorce_rate_by_year)
plt.title('Average Divorce Rate per Year')
plt.show()
```

Based on the barplot above, the divorce rate trends are less drastic compared to birth rates. There was a slight increase between 2000 until 2003 and the trend decreased gradually between 2003 until 2019 with a few non-drastic peaks in between. In 2019 until 2022, the trend started decreasing more steeply compared to past years. Linking the results with marriage rate and birth rate, the decrease in divorce could be due to the decrease in overall marriage.


In relation to the marriage and divorce rates, the correlation of the two with natural growth rate is investigated below.

```python
# 4. The relationship between the natural growth rate and marriage and divorce rate
# Scatter plot and correlation of natural growth rate and marriage rate
plt.figure(figsize=(15, 8))
plt.subplot(1, 2, 1)
sns.scatterplot(x='Natural_growth_rate', y='Marriage_rate', data=kdemo)
plt.title('Natural Growth Rate vs Marriage Rate')
plt.xlabel('Natural Growth Rate')
plt.ylabel('Marriage Rate')

correlation_marriage = kdemo['Natural_growth_rate'].corr(kdemo['Marriage_rate'])
print(f"Correlation between Natural Growth Rate and Marriage Rate: {correlation_marriage:.2f}")

# 4. Scatter plot and correlation of natural growth rate and divorce rate
plt.subplot(1, 2, 2)
sns.scatterplot(x='Natural_growth_rate', y='Divorce_rate', data=kdemo)
plt.title('Natural Growth Rate vs Divorce Rate')
plt.xlabel('Natural Growth Rate')
plt.ylabel('Divorce Rate')

correlation_divorce = kdemo['Natural_growth_rate'].corr(kdemo['Divorce_rate'])
print(f"Correlation between Natural Growth Rate and Divorce Rate: {correlation_divorce:.2f}")

plt.tight_layout()
plt.show()
```

The plot above shows that there is a positive correlation of 0.52 between the natural growth rate and the marriage rate. A higher marriage rate tends to result in a higher birth rate. However the correlation of the natural growth rate with the divorce rate is low at 0.35. This means that the two are most likely independent of each other.

```python
# 5. The average death rate of South Korea between 2000 until the first half of 2022
plt.figure(figsize=(15, 6))
sns.barplot(x='Year', y='Death_rate', data=kdemo_country, errorbar=None)  # ci=None disables error bars
plt.title('Death Rate (2000-2022)')
plt.xlabel('Year')
plt.ylabel('Death Rate')
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.show()
```

The death rate of South Korea seems to have been almost stagnant with a slight decrease between 2000 until 2009 and it started increasing gradually between 2012 until 2021. There is a spike in death rates in 2022 which according to The Korean Herald (2022) was due to the increase of deaths caused by Covid. The steady trend in death rates may indicate that Korea has a low mortality rate because despite the ageing population due to dropping birth rates, the number of deaths per population stays almost the same each year.
